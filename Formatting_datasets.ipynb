### Notebook to define a function to clean and format surveys 
#Remove extra columns from survey dataset
#Add selection column with 1 if selected, 0 if not selected
#Add scale of sounding selection product

```python
#load in libraries and files
import pandas as pd
import numpy as np
import matplotlib as p

survey = pd.read_csv('..//..//..//Documents//Test_sets//Testing_Sample//HI1509//Scale_25000//Edits//HI1509_C.csv')
ss = pd.read_csv('..//..//..//Documents//Test_sets//Testing_Sample//HI1509//Scale_25000//Edits//SS_HI1509_sc25_C.csv')
```

```python
#Count the number of soundings in the selection
len(ss.index)
```

```python
def format_surv(A,B,C,D):
    #where A is the survey, B is the sounding selection, C is the scale of the sounding selection, D is survey name
    
    #Clean up datasets
    A_1= A.loc[:,['X','Y','Depth (m)','fid']]
    B_1= B.loc[:,['X','Y','Depth']]
    
    #Add selection column and scale
    A_1['Selection'] = 0
    A_1['Scale'] = C
    
    #Identify the closest points
    min_pos = []
    errors = []
    
    for i in range(0,len(B_1)):
        x1 = B_1.iloc[i][0]
        y1 = B_1.iloc[i][1]
    
        dist_vect = []
    
        for j in range(0,len(A_1)):
            x2= A_1.iloc[j][0]
            y2= A_1.iloc[j][1]
            
        
                #distance between x and ys
            dist = np.sqrt(np.abs((x2-x1)+(y2-y1)))
            dist_vect.append(dist)
    
        min_value = dist_vect.index(min(dist_vect))
        min_pos.append(min_value)
        errors.append(dist_vect[min_value])

    
    #for all the positions with smallest distance, assume match and change selected to 1
    A_1.iloc[min_pos,4] = 1
    
    #save completed survey file
    filename = "Final" + D + ".csv"
    A_1.to_csv("..//..//..//Documents//Test_sets//Files_for_Disc//testerselectionfiles//" + filename)
    return A_1.head(4)
    
    
```

```python
format_surv(survey,ss,25000,'HI1509')
```
