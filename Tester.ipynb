```python
min_pos = []
errors = []
for i in range(0,len(ss2)):
    x1 = ss2.iloc[i][0]
    y1 = ss2.iloc[i][1]
    #print(x,y)
    
    dist_vect = []
    
    for j in range(0,len(wholefile_d)):
        x2= wholefile_d.iloc[j][0]
        y2= wholefile_d.iloc[j][1]
        
        #distance between x and ys
        dist = np.sqrt(np.abs((x2-x1)+(y2-y1)))
        dist_vect.append(dist)
    
    min_value = dist_vect.index(min(dist_vect))
    min_pos.append(min_value)
    errors.append(dist_vect[min_value])
#print(min_pos)
#print(errors)        
        
```
